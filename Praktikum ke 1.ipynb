### 4) Membaca format CSV

File CSV dapat diuraikan (parsing) ke dalam bentuk list, dictionary, atau dataframe.

Tuliskan code pada cell di bawah ini untuk melakukan parsing csv ke dalam bentuk list

```
#membuat fungsi untukk mengubah csv menjadi List
def csv_list(filename):
  #variabel data yang bertipekan list untuk menampung data csv
  data=[]
  with open(filename) as csv_file :
    #membaca file CSV menggunakan library CSV, dan pemisah "," antar data
    csv_reader =  csv.reader(csv_file, delimiter=",")

    #membaca tiap baris pada file csv
    for row in csv_reader:

      #memindahkan data csv tiap baris ke variabel list data
      data.append(row)

    #menghapus header data(NO,NAMA,TELEPON)
    data.pop(0)
    return data
```

Cek apakah fungsi csv_list dapat berjalan dengan baik

```
#mengurai data dengan memanggil fungsi csv_list
list_kontak = csv_list('kontak.csv')
#menampilkan data yang sudah diuraikan antara header dan isi data
print(list_kontak)
```

Tuliskan code pada cell di bawah ini untuk melakukan parsing csv ke dalam bentuk dictionary

```
#membuat fungsi untuk mengubah file csv menjadi dictionary
def csv_dict(filename):
  #variabel data yang bertipekan list untuk menampung data csv
  data=[]
  with open(filename) as csv_file :
    #membaca file CSV menggunakan library CSV Dictionary Reader
    csv_reader =  csv.DictReader(csv_file)

    #membaca tiap baris pada file csv
    for row in csv_reader:

      #memindahkan data csv tiap baris ke variabel list data
      data.append(row)
    return data
```

Cek apakah fungsi csv_dict dapat berjalan dengan baik




```
#mengurai data dengan memanggil fungsi csv_list
list_kontak = csv_dict('kontak.csv')
#menampilkan data yang sudah diuraikan dalam bentuk dictionary
print(list_kontak)
```
